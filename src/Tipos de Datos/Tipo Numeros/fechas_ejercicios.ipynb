## `Ejercicio 9`
Diferencias entre Fechas y Horas

```python
import datetime
import time
fecha_1 = datetime.datetime.now()
time.sleep(3)
fecha_2 = datetime.datetime.now()

diferencia = fecha_2 - fecha_1
diferencia.seconds
```

```python
# usando deltas

delta = datetime.timedelta(days=7)
fecha_result = fecha_2 + delta
fecha_2.day
fecha_result.day
```
