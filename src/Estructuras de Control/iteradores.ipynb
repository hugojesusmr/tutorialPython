## `Ejercicio 1`

Iterar un Lista 

```python
primos = [2, 3, 5, 7, 11]

iterador = iter(primos)
print(next(iterador))
print(next(iterador))
print(next(iterador))
print(next(iterador))
print(next(iterador))
```

## `Ejercicio 2`
Extraer el contenido de un Iterador con la funcion enumerate

```python
lenguajes = ['Python', 'Java', 'C#', 'PHP', 'C++']

enum_lenguajes = enumerate(lenguajes)
print(next(enum_lenguajes))
print(next(enum_lenguajes))
print(next(enum_lenguajes))
print(next(enum_lenguajes))
print(next(enum_lenguajes))
```

## `Ejercicio 3`

Extraer el contenido de un Iterador con la funcion enumerate asignando un indice

```python

frameworks = ['Spring', 'Laravel', 'Fast API', 'Django']
enum_frame = enumerate(frameworks, 5)
print(next(enum_frame))
print(next(enum_frame))
print(next(enum_frame))
print(next(enum_frame))
```

## `Ejercicio 4`

Filtar un Iterador

```python
numeros = range(10)
impares = list(filter(lambda numero: numero % 2, numeros))
print(impares)
```

## `Ejercicio 5`

Filtrar los numeros pares con el metodo itertools()

```python
import itertools

pares = list(itertools.filterfalse(lambda numero: numero % 2, numeros))
pares
```

## `Ejercicio 6`

Iterar Datos que no cuentan con un iterador

```python
def generar_cuadrado(valor=0):
    while True:
        valor += 1
        yield (valor - 1) ** 2

gen = generar_cuadrado()
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

## `Ejercicio 7`

Modulo itertools()

```python
#funcion accumulate
import itertools

rango = range(0, 10)

acumulador = itertools.accumulate(rango)
# 0 = 0
print(next(acumulador))
# 0 + 1 = 1
print(next(acumulador))
# 1 + 2 = 3 
print(next(acumulador)) 
# 3 + 3 = 6
print(next(acumulador)) 
```

```python
# funcion combinations

rango = range(0,10)
combinaciones = itertools.combinations(rango, 2)
print(next(combinaciones))
print(next(combinaciones))
print(next(combinaciones))
print(next(combinaciones))
```
