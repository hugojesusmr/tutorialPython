## `Ejercicio 2`
Ordenar las Columnas: Id Form, Site Code, Form Name,Project, Region

```python
# --- Definir las columnas a ordenar ---
ordenar_columnas = ['Id Form', 'Site Code', 'Form Name','Project', 'Región']

# --- Crear una lista con el resto de las columnas ---

resto_columnas = [ col for col in odks.columns if col not in ordenar_columnas]

# --- Concatenar las listas a ordenar ---
nuevo_orden = ordenar_columnas + resto_columnas

# --- Aplicar el nuevo orden de columnas ---
odks_ordenado = odks[nuevo_orden]
odks_ordenado.head(2)
```
