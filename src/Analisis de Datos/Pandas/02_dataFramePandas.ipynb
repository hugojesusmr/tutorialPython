## `¿ Qué es un DataFrame ?`

* Un DataFrame es un conjunto de datos con estructura Matricial, formado por Filas y Columnas.
* Se puede Definir como un conjunto de Series que comparten un Indice
* Puede Almacenar Cualquier Tipo de Dato


## `¿ Qué se puede hacer con un DataFrame ?`

* Agruparlos por valores de columnas
* Calcular Sumas, medias, medianas, maximos, minimos, etc.
* Seleccionar Filas, Columnas con ciertos valores
* Concatenar, mezclar combinar varis DataFrames

## `Ejercicio 1`

Primer Forma de construir un DataFrame con datos y columnas 

```python
import pandas as pd

mes_ventas = pd.DataFrame(data=[
    ('Enero', 177643, 12475, 996, 2564),
    ('Febrero', 177643, 54321, 236, 4325),
    ('Marzo', 177643, 98654, 567, 7654),
    ('Abril', 177643, 34986, 123, 9876),
    ('Mayo', 177643, 345123, 890, 7890),
   
], columns=['Mes', 'Ventas', 'Gastos', 'Ingresos', 'Utilidades'])

mes_ventas
```

## `Ejercicio 2`
