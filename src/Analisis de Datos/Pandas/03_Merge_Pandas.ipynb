### `2.- UNION DE DATAFRAMES CON PANDAS(INNER, OUTER, RIGHT, LEFT JOIN)`

Podemos unir o fusionar DataFrames en pandas con python usando la funcion `merge()`, los diferentes argumentos para `merge()` le permiten realizar una unión natural, unión izquierda, unión derecha, unión externa.

### `SINTAXIS: `

`pd.merge(left_df1, right_df2, how="inner", on=None, left_on=None,right_on=None, sort=True, indicator=False)`

### Donde:

* `left_df1, right_df2` : Son los DataFrames

* `how`: left, right, outer, inner indica que tipo de combinación se va a utilizar, el valor predeterminada es inner.<br>
       - `inner:` Mantiene solo las filas que coinciden con los marcos de datos.<br>
       - `outer:` Mantiene todas las Filas de ambos Marcos de Trabajo<br>
       - `left:` Para incluir todas las filas tanto de xy y solo las de y que coinciden<br>
       - `right:` Incluir todas las filas del marcho y y solo las de x que conciden<br>

* `on, left_on, right_on`: Columna que se utilizara como clave
* `sort`: ordena el DF resultane por la columna clave
* `indicator`: añade una columna que indica la procedencia de cada fila

### `Ejercicio 1`
crear dos dataframes


```python
# ---- Data Frame 1 ----

datos_1 = {
    'id':pd.Series([1,2,3,4,5,6]),
    'producto': pd.Series(['Teclado','Radio','Laptop','Televisión','Refrigerador','Leche'])
}

df_1 = pd.DataFrame(datos_1)
df_1
```

```python
# ---- Data Frame 2 ----

datos_2 = {
    'id':pd.Series([2, 4, 6, 7, 8]),
    'precio': pd.Series([100, 200, 300, 400, 500])
}

df_2 = pd.DataFrame(datos_2)
df_2
```

## `Ejercicio 7`

### Inner Join(Union Interna)

Devulve solo las filas en las que la tabla izquierda tienen claves concidentes en la tabla derecha

```python
inner_join = pd.merge(df_1, df_2, on='id', how='inner')
inner_join
```

## `Ejercicio 8`

### Outer Join(Union Externa)

Devuelve todas las filas de ambas tabla, une los registros de la tabla izquierda con la tabla derecha, cuando no hay
coincidentes en ninguna tabla se devolvera NaN

```python
outer_join = pd.merge(df_1, df_2, on='id', how='outer')
outer_join
```

## `Ejercicio 9`

### Left Join(Union Externa Izquierda)

Devuelve todas las filas de la tabla izquierda, y todas las filas con clave coincidentes de la tabla derecha, cuando no hay
coincidentes en la tabla derecha se devolvera NaN

```python
# left join

left_join = pd.merge(df_1,df_2, on='id', how='left')
left_join
```

## `Ejercicio 10`

### Right Join(Union Externa Derecha)

Devuelve todas las filas de la tabla derecha, y todas las filas con clave coincidentes de la tabla izquierda, cuando no hay
coincidentes en la tabla izquierda se devolvera NaN

```python
right_join = pd.merge(df_1, df_2, on='id', how='right')
right_join
```
