## `¿ Qué es una Serie ?`

Una Serie es una estructura de datos que contiene dos vectores:
* El Indice
* Los Datos
* Constructor

## `Ejercicio 1`

Crear un objeto Series de los Meses del Año

```python
import pandas as pd

meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
serie = pd.Series(meses)
serie
```

## `Ejercicio 2`
Agregar un Indice a los meses del Año

```python
# --- Lista de los Meses del Año ---
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
# --- Indice para ca uno de los meses del Año ---
indice  = list(range(1,13))
# --- Crear la Serie con el Indice ---
serie = pd.Series(meses, index=indice)
serie
```

## `Ejercicio 3`

Crear una serie con numeros aleatorios, agregandole como indice una fecha

```python
import random 

fechas = pd.date_range('20250101', periods=12)
datos = [random.randrange(1, 50, 1) for i in range(12)]
serie_fechas = pd.Series(datos, index=fechas)
serie_fechas
```

## `Ejercicio 7`

Crear un directorio

```python
import os

nuevo_directorio = 'nuevo_directorio'
if not os.path.exists(nuevo_directorio):
    os.makedirs(nuevo_directorio)
    print(f"se creo el directorio: {nuevo_directorio}")
else:
    print("Ya existe el directorio")    
```
