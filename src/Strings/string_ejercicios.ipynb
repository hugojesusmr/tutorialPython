## `Ejercicio 9`

Remover el espacio o cualquier otro caracter de una cadena de caracteres

```python
# se eliminan los espacios de izquierda y derecha con el metodo strip()

frase = '    Soy Genial :)    '
frase_sin_espacios = frase.strip()
frase_sin_espacios
```

```python
# se eliminan los espacios de izquierda con el metodo lstrip()
frase = '   Estoy en mi mejor momento'
frase_sin_espacios_izquierdos = frase.lstrip()
frase_sin_espacios_izquierdos
```

```python
# se eliminan los espacios de derecha con el metodo rstrip()
frase = "Puedo lograr lo qeu yome proponga      "
frase_sin_espacios_derechos = frase.rstrip()
frase_sin_espacios_derechos
```

## `Ejercicio 10`

Cambiar entre Mayusculas y Minusculas de una cadena de caracteres

```python
# convertir la frase a mayusculas
frase = "Python es tremendo"
conversion_Mayusculas = frase.upper()
conversion_Mayusculas
```

```python
# convertir la frase a minusculas
frase = "Python es tremendo"
conversion_Minusculas = frase.lower()
conversion_Minusculas
```

```python
# convertir la primer letra a mayuscula
frase = "python es tremendo"
conversion_PrimerLetra = frase.capitalize()
conversion_PrimerLetra
```

### Función `startwith()`

 devuelve un valor booleano True o False si coincide que la cadena inicie con el criterio enviado por parámetros en la función.

```python
'Hola Mundo'.startswith('Ho')
```

### Función `endswith()`

Devuelve un valor booleano True o False so coincide con la cadena 

```python
'Hola Mundo'.endswith('do')
```

### Función `find()`

Devuelve un valor númerico `0` si encuentra el criterio y `-1` si no coincide

```python
'Esto es un Test'.find('esto')
```

```python
"Hola Hola".find('Hol')
```

### Función `isalnum()`

Devuelve un valor booleano `True` o `False` si la cadena contiene caracteres alfanumericos

```python
"12345".isalnum()
```

### Función `isalpha()`

Devuelve un valor `True` o `False` si la cadena contiene caracteres alfabeticos.

```python
'Python'.isalpha()
```
