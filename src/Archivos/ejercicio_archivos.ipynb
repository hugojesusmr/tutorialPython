## `Ejercicio 2`
Extraer el contenido de un Iterador con la funcion enumerate

```python
lenguajes = ['Python', 'Java', 'C#', 'PHP', 'C++']

enum_lenguajes = enumerate(lenguajes)
print(next(enum_lenguajes))
print(next(enum_lenguajes))
print(next(enum_lenguajes))
print(next(enum_lenguajes))
print(next(enum_lenguajes))
```

## `Ejercicio 3`

Extraer el contenido de un Iterador con la funcion enumerate asignando un indice

```python

frameworks = ['Spring', 'Laravel', 'Fast API', 'Django']
enum_frame = enumerate(frameworks, 5)
print(next(enum_frame))
print(next(enum_frame))
print(next(enum_frame))
print(next(enum_frame))
```

## `Ejercicio 7`

Crear un directorio

```python
import os

nuevo_directorio = 'nuevo_directorio'
if not os.path.exists(nuevo_directorio):
    os.makedirs(nuevo_directorio)
    print(f"se creo el directorio: {nuevo_directorio}")
else:
    print("Ya existe el directorio")    
```

## `Ejercicio 8`

Obtener Metadatos de una carpeta

```python
import os

directorio_actual = os.getcwd()
metadatos = os.stat(directorio_actual)
print(metadatos)
```
