# Compresión de Listas

* Permiten crear listas de elementos en una sola linea de codigo <br>, 
  cada posición puede ser un tipo de datos distinto.

## `Sintaxis:` lista = [expresión for elemento in iterable]

### Crear una Lista con los cuadrados de los primeros 5 números

```python
cuadrados = [i**2 for i in range(5)]
cuadrados
```

## LLamada a una Función listando los cuadrados de los primeros 5 números

```python
def cuadrados(num):
    return num**2

num_list = [cuadrados(i) for i in range(1,6)]
num_list
```

## Añadiendo condicionales

Podemos realizar operaciones sobre elementos sólo si se cumple determida condición

### Sintaxis: `lista = [expresion for elemento in iterable id condicion ]` 

## Ejercicio

Saber el número de erres que tiene

```python
frase = "El perro de san roque no tiene rabo"

erres = [i for i in frase if i == 'r']
erres
```

## Set Compresions

Lo set no permite duplicados, si intentamos añadir un elemento que ya existe en el set, simplemente no se añade

```python
frase = "El perro de san roque no tiene rabo"

erres = {i for i in frase if i =='r'}
erres
```

## Diccionario de Compresión

Extiende una lista agregando una lista al final

```python
lista1 = ['nombre', 'edad', 'región']
lista2 = ['Pelayo', 30, 'Asturias']

my_dict = {i:j for i,j in zip(lista1, lista2)}
my_dict
```

## Función `index()`

Este método recibe un elemento como argumento, y devuelve el índice de su primera aparición en la lista.

```python
# numbers.index(elemento)
numbers = [1,2,3,4,5]

#En el indice 2 de la lista esta el elemento 3
numbers.index(3)
```

## Función `insert()`

Este método inserta un elemento en el índice (i) en la lista.

```python
letras = ['a','b','b','f']
letras.insert(2,'c')
letras
```

## Función `pop()`

Este método devuelve el ultimo elemento de la lista y lo borra.

```python
elements = [1,3,5,7]
elements.pop()
elements
```

Opcionalmente puede recibir un argumento numérico, que funciona como índice del elemento (por defecto, -1)

```python
elements.pop(1)
elements
```

## Función `remove()`

Este método recibe como argumento un elemento, y borra su primera aparición en la lista.

```python
try:
    elements.remove(1)
except ValueError as e:
    print(e)    
```

## Función `reverse()`

Este método invierte el orden de los elementos de una lista

```python
n = [1,2,3,4,5]
n.reverse()
n
```

## Función `sort()`

Este método ordena los elementos de una lista.

```python
versiones_plone = [4, 2.5, 5, 3.6, 2.1, 6]
versiones_plone.sort()
versiones_plone
```
